```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
#pick school
school_dbn = '30Q580'
school_df = df[df['School DBN']== school_dbn] #this dataframe only has the selected DBN's data
```

```python
school_df
```

```python
#select feature
feature = 'Mean Score'

#change mean score column to a number
df[feature]= pd.to_numeric(df[feature], errors='coerce')
#to_numeric function changes non-numeric values in selected dataframe to numbers. if string is "100" then it will change value to 100
#if value is "n/a" or "unknown" use error parameter (coerce changes value to NaN, raise will raise an error, and ignore means no convesion)

df = df.dropna(subset=[feature]) #drop all missing values in the selected feature
```

```python
df[feature]
```

```python
df.describe()
```

```python
school_df[feature]= pd.to_numeric(df[feature], errors='coerce')

school_df = school_df.dropna(subset=[feature])
```

```python
school_df
```

```python
queens_df=df[df['School DBN'].str.contains('Q')]
#filter for schools with Q in the School DBN
```

```python
sns.barplot(x=[school_dbn, 'Queens School'], y=[school_df[feature].mean(), queens_df[feature].mean()], palette=['green', 'red'])
plt.title('BSGE Average Regents Scores v. Queens Schools Average Regents Scores')
plt.ylabel('Test Score')
```

```python
# From the above visualization we can see that school 30Q580 mean score is higher than the mean score of the schools in Queens.
# This visualization shows the averages of all the mean scores for the selected school and the averages of all mean scores for queens schools.
# So, we can say that the selected school is getting above average scores on their regents exams, but it would be better to check by year and specific regents exam.
```

```python
regents_df = df[['School DBN','School Name','Regents Exam', 'Year', 'Total Tested','Mean Score']]
#new dataframe with specific columns
```

```python
chem_2015_df = regents_df[(regents_df['Regents Exam'] == 'Physical Settings/Chemistry') & (regents_df['Year']==2015)]
#dataframe that contains chem exam in 2015
```

```python
school_chem_2015 = chem_2015_df[chem_2015_df['School DBN']==school_dbn]
queens_chem_2015 = chem_2015_df[chem_2015_df['School DBN'].str.contains('Q')]
```

```python
school_chem_2015.describe()
```

```python
queens_chem_2015.describe()
```

```python
sns.barplot(x= [school_dbn, 'Queens Schools'], y= [school_chem_2015['Mean Score'].mean(), queens_chem_2015['Mean Score'].mean()], palette=['blue', 'orange'])
plt.title('BSGE Average Chemistry Scores v. Queens Schools Average Chemistry Scores in 2015')
plt.ylabel('Test Score')
```

```python
sns.distplot(school_chem_2015['Mean Score'])
```

```python
sns.distplot(queens_chem_2015['Mean Score'])
```

```python
# From the above 3 visualizations, we can say that BSGE average Chemistry regents score was above average compared to all Queens Schools in 2015.
# Most Students in BSGE scored between 82 and 85 while in Queens schools most students scored between 60 and 75.
```

```python
sns.barplot(x= [school_dbn, 'All Schools'], y= [school_chem_2015['Mean Score'].mean(), chem_2015_df['Mean Score'].mean()], palette=['blue', 'orange'])
plt.title('BSGE Average Chemistry Scores v. All Schools Average Chemistry Scores in 2015')
plt.ylabel('Test Score')
```
