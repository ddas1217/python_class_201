```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
analysis_df = df[['race', 'number_diagnoses']]
```

```python
analysis_df
```

```python
analysis_df= analysis_df[(analysis_df['race'].notna()) & (analysis_df['race'] != '?') &(analysis_df['number_diagnoses'].notna())]
#cleaning the selected columns getting rid of NaN values or '?'
```

```python
analysis_df
```

```python
analysis_df.describe()
```

```python
race_diagnoses_avg = analysis_df.groupby('race')['number_diagnoses'].mean().reset_index()
#group by race and average number of diagnoses. have to reset index because race became the index if you don't use reset_index function see below.
```

```python
race_diagnoses_avg
```

```python
palette=sns.color_palette('Set2') #Set2 is one of many color palettes in seaborn library.
ax=sns.barplot(x=race_diagnoses_avg['race'], y=race_diagnoses_avg['number_diagnoses'], data= race_diagnoses_avg, palette=palette)
plt.title('Average Number of Diagnoses based on Race')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
ax.bar_label(ax.containers[4])
```

```python
analysis_df2 = df[['race', 'num_lab_procedures']]
```

```python
analysis_df2
```

```python
analysis_df2= analysis_df2[(analysis_df2['race'].notna()) & (analysis_df2['race'] != '?') &(analysis_df2['num_lab_procedures'].notna())]
#cleaning the selected columns getting rid of NaN values or '?'
```

```python
race_procedures_avg = analysis_df2.groupby('race')['num_lab_procedures'].mean().reset_index()
```

```python
race_procedures_avg
```

```python
palette=sns.color_palette('pastel') #pastel is one of many color palettes in seaborn library.
ax = sns.barplot(x=race_procedures_avg['race'], y=race_procedures_avg['num_lab_procedures'], data= race_procedures_avg, palette=palette)
plt.title('Average Number of procedures based on Race')
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[2])
ax.bar_label(ax.containers[3])
ax.bar_label(ax.containers[4])
```

Based off the two visualizations we see that Caucasians have the most average diagnoses even though they have the third most lab procedures done. Depending on what these diagnoses are it could mean that caucasians are more unhealthy compared to african americans who had the most lab procedures done and had the third most diagnoses.
